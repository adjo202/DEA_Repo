```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#IMPORTING DATASET
file_path = 'DEA_Dataset.xlsx'

defenders_df = pd.read_excel(file_path, sheet_name='Defenders')
midfielders_df = pd.read_excel(file_path, sheet_name='Midfielders')
forwards_df = pd.read_excel(file_path, sheet_name='Forwards')
```

# DEFENDERS VISUALISATIONS

```python
plt.figure(figsize=(12, 8))
sns.set(style='darkgrid')
plot = sns.scatterplot(x='Tackles+Interceptions per 90', y='Clearances per 90', data=defenders_df, s=200, color="red", edgecolor='black')

# Title and labels
plt.title('Defender Performance: Tackles+Interceptions vs. Clearances per 90', fontsize=16)
plt.xlabel('Tackles+Interceptions per 90')
plt.ylabel('Clearances per 90')


for line in range(0, defenders_df.shape[0]):
    plot.text(defenders_df['Tackles+Interceptions per 90'][line]+0.07,
              defenders_df['Clearances per 90'][line],
              defenders_df.index[line] + 1,
              horizontalalignment='left',
              size='medium',
              color='black')

plt.axis('equal')
plt.grid(True, color='white')
plt.show()
```

```python
# Set figure size and background style
plt.figure(figsize=(12, 8))
sns.set(style='darkgrid')

defenders_df['Goals Conceded per 90 inverted'] = defenders_df['Goals Conceded per 90 inverted'] * -1

plot = sns.scatterplot(x='Total Duel Won per 90', y='Goals Conceded per 90 inverted', data=defenders_df,
                       s=200, color="red", edgecolor='black')

plt.title('Defender Performance: Total Duels Won vs. Goals Conceded per 90', fontsize=16)
plt.xlabel('Total Duel Won per 90', fontsize=12)
plt.ylabel('Goals Conceded per 90 (Inverted)', fontsize=12)

for line in range(0, defenders_df.shape[0]):
    plot.text(defenders_df['Total Duel Won per 90'][line]+0.05,
              defenders_df['Goals Conceded per 90 inverted'][line],
              defenders_df.index[line] + 1,
              horizontalalignment='left',
              size='medium',
              color='black',
              weight='semibold')

plt.axis('equal')
plt.grid(True, color='white')
plt.show()
```

# MIDFIELDERS VISUALISATIONS

```python
plt.figure(figsize=(12, 8))
sns.set(style='darkgrid')

plot = sns.scatterplot(x='Passes Completed per 90', y='Key Passing per 90', data=midfielders_df,
                       s=200, color="red", edgecolor='black')

plt.title('Midfielder Performance: Passes Completed vs. Key Passes per 90', fontsize=16)
plt.xlabel('Passes Completed per 90', fontsize=12)
plt.ylabel('Key Passes per 90', fontsize=12)

for line in range(0, midfielders_df.shape[0]):
    plot.text(midfielders_df['Passes Completed per 90'][line] + 0.5,
              midfielders_df['Key Passing per 90'][line],
              str(line + 1),
              horizontalalignment='left',
              size='medium',
              color='black',
              weight='semibold')

plt.grid(True, color='white')
plt.show()
```

# FORWARDS VISUALISATIONS

```python
plt.figure(figsize=(12, 8))
sns.set(style='darkgrid')

plot = sns.scatterplot(x='Goals per 90', y='Excpected Goals per 90', data=forwards_df,
                       s=200, color="red", edgecolor='black')

plt.title('Forward Performance: Goals vs. Expected Goals per 90', fontsize=16)
plt.xlabel('Goals per 90', fontsize=12)
plt.ylabel('Expected Goals per 90', fontsize=12)

for line in range(0, forwards_df.shape[0]):
    plot.text(forwards_df['Goals per 90'][line] + 0.006,
              forwards_df['Excpected Goals per 90'][line],
              str(line + 1),
              horizontalalignment='left',
              size='medium',
              color='black',
              weight='semibold')

plt.grid(True, color='white')
plt.show()
```
